# Outer Product

So far we have learned two new operations we can do with bras and kets, namely, tensor and inner product.

$$ |0\rangle \otimes |1\rangle = |01 \rangle $$
$$ \langle 1 | 1 \rangle = 1 $$

Now we define a new operation called outer product and it is defined as:

$$
|0\rangle\langle0|= \begin{pmatrix} 1 \\ 0 \end{pmatrix} \begin{pmatrix} 1 & 0 \end{pmatrix} = \begin{pmatrix} 1&0 \\ 0&0 \end{pmatrix},\\
|0\rangle\langle1| = \begin{pmatrix} 1 \\ 0 \end{pmatrix} \begin{pmatrix} 0 & 1 \end{pmatrix} = \begin{pmatrix} 0&1 \\ 0&0 \end{pmatrix},\\
|1\rangle\langle0| = \begin{pmatrix} 0 \\ 1 \end{pmatrix} \begin{pmatrix} 1 & 0 \end{pmatrix} = \begin{pmatrix} 0&0 \\ 1&0 \end{pmatrix},\\
|1\rangle\langle1| = \begin{pmatrix} 0 \\ 1 \end{pmatrix} \begin{pmatrix} 0 & 1 \end{pmatrix} = \begin{pmatrix} 0&0 \\ 0&1 \end{pmatrix}.\\
$$

$$ |\psi \rangle = a_0 |0 \rangle + a_1 |1 \rangle =  \begin{pmatrix} a_0 \\ a_1 \end{pmatrix} $$
$$ |\phi \rangle = b_0 |0 \rangle + b_1 |1 \rangle =  \begin{pmatrix} b_0 \\ b_1 \end{pmatrix} $$

a. $$|\psi \rangle \langle \phi | = \begin{pmatrix} a_0 \\ a_1 \end{pmatrix}  \begin{pmatrix} b_0&b_1 \end{pmatrix} = \begin{pmatrix} a_0b_0&a_0b_1 \\ a_1b_0 & a_1b_1 \end{pmatrix}\\ $$

b. $$|\phi \rangle \langle \psi | = \begin{pmatrix} b_0 \\ b_1 \end{pmatrix}  \begin{pmatrix} a_0&a_1 \end{pmatrix} = \begin{pmatrix} b_0a_0&b_0a_1 \\ b_1a_0 & b_1a_1 \end{pmatrix}\\ $$

It is quite obvious that given an arbitrary operator $M$ and its matrix representation, we can write it down in terms of outer products of $|0\rangle$ and $|1\rangle$ as follows:

$$
M= \begin{pmatrix} m_{0,0}&m_{0,1} \\ m_{1,0}&m_{1,1} \end{pmatrix} = m_{0,0} |0\rangle\langle0|+ m_{0,1} |0\rangle\langle1|+ m_{1,0} |1\rangle\langle0|+ m_{1,1} |1\rangle\langle1|
$$

This gives us a clue as to how we can construct a quantum circuit for a given unitary transformation. However, note that $|0\rangle \langle 1|$ is not a unitary operator, hence not a valid quantum gate. But luckily we can solve this problem very easily by noting the following relations:

$$
\frac{1+Z}{2} = \frac{1}{2}\left[ \begin{pmatrix} 1&0 \\0&1 \end{pmatrix}+\begin{pmatrix} 1&0 \\0&-1 \end{pmatrix}\right] = |0\rangle\langle0|,\\\frac{1-Z}{2} = \frac{1}{2}\left[ \begin{pmatrix} 1&0 \\0&1 \end{pmatrix}-\begin{pmatrix} 1&0 \\0&-1 \end{pmatrix}\right] = |1\rangle\langle1|
$$

$$
|0\rangle\langle1| = |0\rangle\langle0|X = \frac{1}{2}(1+Z)~X = \frac{X+iY}{2},\\\\
|1\rangle\langle0| = X|0\rangle\langle0| = X~\frac{1}{2}(1+Z) = \frac{X-iY}{2}.
$$

# Measuring Time Complexity

 Now, before we move on to our first quantum algorithm we must discuss what time complexity is. As you can probably guess, when you want to solve a problem using computers, there are a lot of ways to go about it. However, some solutions might be more efficient than others in terms of their memory or time usage. We are mainly interested in time complexity of an algorithm. What we want to understand is how does our algorithm scale as the input size gets bigger. For this purpose, we are neglecting the constant time operations most of the time and focus on the repetetive parts of the algorithm.
 
 Most of the time we represent the complexity of an algorithm using big-O notation which is  we define as follows:
 
 Let f and g two real and positive valued functions. If there exists a positive number $C$ and a real value $x_0$ such that $f(x) \leq C g(x) \text{ } \forall \text{ } x \geq x_0 $ we say $f(x) = O ( g(x) ) $.
 
 
 
 
 
 So basically, $g(x)$ is an upper bound for $f(x)$. 
 
 This notation saves us from a lot of trouble. For example, given a $f(x) = x^4 + 20000 x^3 + 199 x^2 + 10^9 $, we can immediately write $ f(x) = O(x^4)$. Even though the $x^3$ term and the constant term is quite large in $f(x)$, we can be sure that we can always find $C$ and $x_0$ such that $f(x) \leq C g(x)$ for all $x \geq x_0$.

In Deutch's problem, we are given an oracle that implements a binary function $f: \{0,1\} \rightarrow \{0, 1 \}$, and we are asked to determine whether this function is $\textit{constant}$ or $\textit{balanced}$. We say a binary function $f(x)$ is constant if $f(0) = f(1)$ and it is balanced if $f(0) \neq f(1)$. In quantum computing words, we have the following gate:

# Exercise 7

Let's say $f$ takes $x$ to its complement. That is, $f(0) = 1$ and $f(1) = 0$. Is $f$ balanced or constant?

# Answer

Balanced because $ f(0) \neq f(1) $ 

So in order to solve this problem without quantum computers, we have to check oracle's output for 0 and 1, and decide accordingly. Hence, best classical algorithm we can come up with uses the oracle 2 times. But we can do better with a quantum computer:



Let's write down each state explicitly:

$$ |\phi_0 \rangle = |01\rangle $$
$$ |\phi_1 \rangle = \frac{1}{2} \big( |00\rangle -  |01\rangle +  |10\rangle -  |11\rangle \big)$$
$$ |\phi_2 \rangle =  \frac{1}{\sqrt{2}} \big( (-1)^{f(0)} |0\rangle + (-1)^{f(1)} |1\rangle \big)\otimes |- \rangle $$

So if $f$ is balanced, there will be no phase difference between the coefficients of $|0\rangle$ and $|1\rangle$ on the first qubit, hence it will be a $|+\rangle$ state. If the function is balanced, then the signs of the coefficients of $|0\rangle$ and $|1\rangle$ will be different and the first qubit will be in $|-\rangle$ state. Hence applying a Hadamard will either transform the first qubit to $|0\rangle$ if $f$ is constant, $|1\rangle$ if $f$ is balanced.
